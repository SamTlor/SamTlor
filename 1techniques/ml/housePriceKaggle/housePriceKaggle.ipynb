```python
import pandas as pd
training = pd.read_csv("train.csv")
training = training.drop(["Id"], axis = "columns")

#Get shape and head
print("(rows, columns)")
print(training.shape)
training.head()
```

```python
x = training.drop("SalePrice", axis = 1)
y = training.SalePrice

categorical_columns = x.select_dtypes(include=['object']).columns
x = pd.get_dummies(x, columns=categorical_columns, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import accuracy_score

# Split data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Initialize and train the Decision Tree
dt_model = DecisionTreeClassifier(random_state=42)  # Use DecisionTreeRegressor for regression

# TODO: this breaks because x_train has nan values so we'll need to use imputation
dt_model.fit(x_train, y_train)

# Predict on test set
y_pred = dt_model.predict(x_test)

# Evaluate the model (accuracy for classification, or use R2 for regression)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
```

```python
# To see the structure of the tree:
from sklearn import tree
tree.plot_tree(dt_model)
```
