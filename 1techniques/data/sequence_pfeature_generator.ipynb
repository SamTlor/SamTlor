```python
# from pfeature_comp import main as pfeature_comp
# import sys
# import runpy
import pandas as pd
```

```python
df = pd.read_csv('sequences_training.txt', header=None)
# get rid of the last column
df = df.iloc[:, :-1]
```

```python
# get the last column
df_label = pd.read_csv('sequences_training.txt', header=None)
df_label = df_label.iloc[:, -1]
```

```python
# replace the columns that includes B,J,O,U,Z or X with M and record the index to a list
index = []
for i in range(df.shape[0]):
    for j in range(df.shape[1]):
        if df.iloc[i, j].find('B') != -1 or df.iloc[i, j].find('J') != -1 or df.iloc[i, j].find('O') != -1 or df.iloc[i, j].find('U') != -1 or df.iloc[i, j].find('Z') != -1 or df.iloc[i, j].find('X') != -1:
            df.iloc[i, j] = 'M'
            index.append((i, j))
```

```python
# create a column and make the value 1 for the index in the list
df['UNNTRL'] = 0
for i in index:
    df.iloc[i[0], -1] = 1
```

```python
# write the sequences to a text file
with open('sequences.txt', 'w') as f:
    for index, row in df.iterrows():
        f.write(row[0] + '\n')
```

```python
# open the csv files with any name in the output folder
df_aac = pd.read_csv('output/AAC.csv')
df_atc = pd.read_csv('output/ATC.csv')
df_btc = pd.read_csv('output/BTC.csv')
df_dpc = pd.read_csv('output/DPC.csv')
df_pcp = pd.read_csv('output/PCP.csv')
df_rri = pd.read_csv('output/RRI.csv')
df_ddr = pd.read_csv('output/DDR.csv')
df_sep = pd.read_csv('output/SEP.csv')
df_ser = pd.read_csv('output/SER.csv')
df_spc = pd.read_csv('output/SPC.csv')
```

```python
# combine all dataframes
df_combined = pd.concat([df_aac, df_atc, df_btc, df_dpc,
                        df_pcp, df_rri, df_ddr, df_sep, df_ser, df_spc], axis=1)
```

```python
# add the label column
df_combined['class'] = df_label
```

```python
# export the dataframe to a csv file
df_combined.to_csv('combined.csv', index=False)
```
