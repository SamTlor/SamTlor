```python
import requests
from bs4 import BeautifulSoup

# URL of the published Google Doc
url = (
    "https://docs.google.com/document/d/e/"
    "2PACX-1vSFTq6KR8ER5h9_bFVliDvYBntK6Wv8L7x6hLp2Sm58Zkhpo7Vsba9BmC82wcy8WoR3Q47J-brCiH3c/pub"
)

# Fetch document
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
text = soup.get_text(separator="\n")

# Extract section after "y-coordinate"
after = text.split("y-coordinate", 1)[1].strip()

# Break into lines; each group of 3 lines is one record (x, value, y)
lines = after.splitlines()
rows = [lines[i:i+3] for i in range(0, len(lines), 3)]

# Extract first and last columns (x and y)
x_values = [int(row[0]) for row in rows]
y_values = [int(row[2]) for row in rows]

# Determine grid size
xmax = max(x_values)
ymax = max(y_values)

# Initialize empty grid
grid = [[" " for _ in range(xmax + 1)] for _ in range(ymax + 1)]

# Fill grid with values
for x_str, val, y_str in rows:
    x = int(x_str)
    y = ymax - int(y_str)   # invert y-axis so higher y is at top
    grid[y][x] = val

# Print final grid
for row in grid:
    print(row)
```

```python
def create_staircase(nums):
  step = 1
  subsets = []
  while len(nums) != 0:
    if len(nums) >= step:
      subsets.append(nums[0:step])
      nums = nums[step:]
      step += 1
    else:
      return False
      
  return subsets



nums = [1, 2, 3, 4, 5, 6, 7]
print(create_staircase(nums))
```
